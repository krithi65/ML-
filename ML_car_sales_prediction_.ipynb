```python
df_new = pd.read_csv(
    io.BytesIO(uploaded_file['Car_sales.csv']),
    parse_dates=['Manufacturer'],
).drop('Model', axis=1)
```

```python
df_new
```

```python
import numpy as np

df_new['Time'] = np.arange(len(df_new.index))

df_new.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn-whitegrid")
plt.rc(
    "figure",
    autolayout=True,
    figsize=(11, 4),
    titlesize=18,
    titleweight='bold',
)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=16,
    titlepad=10,
)

fig, ax = plt.subplots()
ax.plot('Time', 'Sales_in_thousands', data=df_new, color='0.75')
ax = sns.regplot(x='Time', y='Sales_in_thousands', data=df_new, ci=None, scatter_kws=dict(color='0.25'))
ax.set_title('Time Plot of Sales_in_thousands');
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics

X = df_new.loc[:, ['Time']]  # features
y = df_new.loc[:, 'Sales_in_thousands']  # target

# Train the model
model = LinearRegression()
model.fit(X, y)

# Store the fitted values as a time series with the same time index as
# the training data
y_pred = pd.Series(model.predict(X), index=X.index)
print(model.coef_,model.intercept_)
print(metrics.r2_score(y,y_pred))
```
